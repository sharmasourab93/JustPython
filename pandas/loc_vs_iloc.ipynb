# .loc Vs .iloc 

#### This notebook is written with an intent to understand and learn the differences between loc method & iloc method used in a DataFrame.

```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('data/survey_results_public.csv')
```

```python
df.head()
```

```python
df.columns.values
```

---

### Using .loc

_loc()_ is a label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike _iloc()_. _loc()_ can accept the boolean data unlike _iloc()_.

```python
ldf = df.copy()
```

```python
help(ldf.loc)
```

```python
ldf.loc[(ldf.Age >= 30) & (ldf.Country == 'United Kingdom'), ['MainBranch', 'CurrencyDesc']]
```

```python
ldf.loc[(ldf.Age > 30), ['YearsCode']]
```

```python
ldf.loc[1]
```

```python
ldf.loc[1, 'MainBranch']
```

```python
ldf.loc[[1, 2]]
```

```python
ldf.loc[[1, 2], 'Country']
```

```python
ldf.loc[[1, 2], ['Country', 'CurrencyDesc', 'MainBranch']]
```

```python
ldf.loc[0:5]
```

```python
ldf.loc[0:5, 'MainBranch':'CurrencyDesc']
```

---

### Using iloc

_iloc_ is an indexed based selecting method which means that we have to pass integer index in the method to select speicific row/column. This method does not include the last element of the range passed in it unlike _loc()_. 
_iloc()_ does not accept the boolean data unlike _loc()_. 

```python
idf = df.copy()
```

```python
help(idf.iloc)
```

```python
# selecting rows from 1 to 4 and columns from 3 to 7
idf.iloc[:5, 3: 7]
```

```python
# idf.loc[:, 'Country'] is equivalent to below
idf.iloc[:, 8]
```

```python
# Prints the 8th row's Country columns' Value i.e. Tunisia
idf.loc[:, 'Country'].iloc[8]
```

```python
# Rows [0, 2, 4, 6, 7] & columns [1, 2, 3, 4, 5]
idf.iloc[[0, 2, 4, 6, 7], [1, 2, 3, 4, 5]]
```

```python
country = idf.columns.get_loc('Country')
country, type(country)
```

```python
idf.loc[idf.index[0:10], 'Country']
```

```python
value = df['Age'] < 30
```

```python
idf.iloc[value.values, [1, 3]]
```

#### Using Filters

```python
high_salary = (df['ConvertedComp'] > 70000)
df.loc[high_salary]
```

```python
countries = ["United States", "India", "United Kingdom", "Germany", "Canada"]
filter_ = df["Country"].isin(countries)
df.loc[filter_, "Country"]
```

```python
df["LanguageWorkedWith"]
```

```python
str_filter = df['LanguageWorkedWith'].str.contains('Python', na=False)
df.loc[str_filter, 'LanguageWorkedWith']
```
